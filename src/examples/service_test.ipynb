```python
import io
import requests
import pandas as pd
```

```python
server_url: str = "http://127.0.0.1:8000"
data_url: str = (
    "https://raw.githubusercontent.com/hse-mlds/ml/main/hometasks/HT1/cars_train.csv"
)
```

```python
# test /predict_item

path = "/predict_item"
df = pd.read_csv(data_url)
item = df.sample(1).to_dict(orient="records")
resp = requests.post(server_url + path, json=item[0])
resp.content.decode("utf-8")
```

```python
# test /predict_items

path = "/predict_items"
df = pd.read_csv(data_url)
df.sample(10).drop(columns="selling_price").to_csv(
    buffer := io.StringIO(), index=False
)
resp = requests.post(server_url + path, files={"csv": buffer.getvalue()})
pd.read_csv(io.BytesIO(resp.content))
```
